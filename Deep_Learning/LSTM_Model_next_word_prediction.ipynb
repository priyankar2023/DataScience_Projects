## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

## Reading input file

```python
with open("C:\\Users\\pkuma262\\OneDrive - JNJ\\Desktop\\Anaconda_Jupyter_NoteBook\\LSTM\\IndiaUS.txt", 'r', encoding='utf-8') as myfile:
    mytext = myfile.read()
```

```python
mytext
```

## Tokenizing the text and assigning unique number to each words

```python
mytokenizer = Tokenizer()
mytokenizer.fit_on_texts([mytext])
total_words = len(mytokenizer.word_index) + 1
```

```python
total_words
```

```python
mytokenizer.word_index
```

## Preparing sequences

```python
my_input_sequences = []

for x in mytext.split("\n"):
#     print(x)
    token_list = mytokenizer.texts_to_sequences([x])[0]
#     print(token_list)
    for i in range(1, len(token_list)):
        my_n_gram_sequence = token_list[: i+1]
#         print(my_n_gram_sequence)
        my_input_sequences.append(my_n_gram_sequence)
```

```python
my_input_sequences
```

## Preparing training data

```python
max_sequence_len = max([len(x) for x in my_input_sequences])

# pad_sequences - To make sure all the length are same (my_input_sequences) for training purpose
input_sequences = np.array(pad_sequences(my_input_sequences, 
                                         maxlen = max_sequence_len, 
                                         padding = "pre"))
```

```python
input_sequences
```

## Breaking input_sequences into features and target

```python
x = input_sequences[:, :-1]

y = input_sequences[:, -1]
```

```python
x
```

```python
y
```

## Converting target to OneHot vectors

```python
y = np.array(tf.keras.utils.to_categorical(y, num_classes = total_words))

y
```

## Building LSTM Model

```python
# Model type is Sequential
model = Sequential()
# Embedding layer
model.add(Embedding(total_words, 100, input_length = max_sequence_len - 1))
# Building LSTM model with 150 units (neurons)
model.add(LSTM(150))
# Creating Dense layer.
# Since out target is categorical (multiclass), hence using Softmax activation
model.add(Dense(total_words, activation = "softmax")) 
print(model.summary())
```

## Model Compilation

```python
model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

## Training model

```python
model.fit(x, y, epochs = 100, verbose = 1)
```

## Predictions

```python
# input_text is our input and we will be predicting predict_next_words words 

input_text = "Joe biden" 
predict_next_words= 6

# input_text = "India" 
# predict_next_words= 2

for _ in range(predict_next_words):
    token_list = mytokenizer.texts_to_sequences([input_text])[0]
    print("Token Number for {}: ".format(input_text), token_list)
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=-1)
    output_word = ""
    for word, index in mytokenizer.word_index.items():
        if index == predicted:
            output_word = word
            break
    input_text += " " + output_word

print(input_text)
```
